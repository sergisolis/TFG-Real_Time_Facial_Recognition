```python
import matplotlib.pyplot as plt
import pickle
import os

plt.figure(figsize=(12, 6))

for filename in os.listdir('triplet_loss_history'):  
    if filename.endswith(".pkl"):
        with open('triplet_loss_history/' + filename, 'rb') as f:
            history = pickle.load(f)
        plt.plot(history['train_loss'], label=f'{filename} training')
        plt.plot(history['test_loss'], label=f'{filename} validation')


plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper right')
plt.show()
```
